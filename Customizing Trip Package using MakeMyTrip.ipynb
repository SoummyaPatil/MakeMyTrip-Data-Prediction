```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("MMT-2.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
df = df.rename(columns={'Uniq Id': 'Uniq_Id', 'Package Name': 'Package_Name','Package Type': 'Package_Type','Places Covered': 'Places_Covered','Travel Date': 'Travel_Date','Start City': 'Start_City','Price Per Two Persons': 'Price_Per_Two_Persons','Per Person Price': 'Per_Person_Price'})
```

```python
df = df.drop(columns =['Crawl Timestamp','Meals','Flight Stops','Initial Payment For Booking','Unnamed: 22','Unnamed: 23','Date Change Rules'])
```

```python
df.dtypes
```

```python
df = df[df["Package_Type"].str.contains("Standard|Deluxe|Premium|Budget|Luxury")==True]
```

```python
df = df[df["Hotel_Details" and "Airline"].str.contains("NA")==False]
```

```python
df.Travel_Date=pd.to_datetime(df.Travel_Date)
```

```python
df
```

Most preferred Package Type

```python
df.Package_Type.value_counts().sort_index().plot.bar(title="Most preferred Package Type",figsize=(6,6));
```

```python
sum_per_packageType = df.groupby('Package_Type').mean()
```

```python
sum_per_packageType
```

```python
sum_per_packageType.plot.bar(figsize=(8,8));
```

Most Preferred Location

```python
sum_per_destination = df.groupby('Destination').mean()
```

```python
sum_per_destination
```

```python
top_5_destinations = pd.DataFrame(df['Destination'].value_counts().head())
```

```python
plt.figure(figsize=(8,8))
sns.barplot(data=df, x=top_5_destinations['Destination'],y=top_5_destinations.index)
plt.title("Top 5 Destinations")
plt.show()
```
